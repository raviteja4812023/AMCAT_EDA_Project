```python
def num_analysis(num_data):

    fig, axes = plt.subplots(1, 3, figsize=(8, 3), constrained_layout=True)
    fig.suptitle("Continuous Distribution Plot")

    axes[0].set_title("Histogram Plot")
    sns.histplot(num_data, ax=axes[0])

    axes[1].set_title("KDE Plot")
    sns.kdeplot(num_data, fill=True, ax=axes[1])

    axes[2].set_title("Box Plot")
    sns.boxplot(num_data, ax=axes[2])

    plt.show()
```

**correlation between college GPA and AMCAT scores**

```python
# Select relevant columns
test_scores = ['English', 'Logical', 'Quant', 'ComputerProgramming', 'ElectronicsAndSemicon',
               'ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg']

# Create a new DataFrame with test scores
amcat_analysis = amcat[test_scores].copy()

# Calculate average AMCAT score and percentage
amcat_analysis['avg_amcat_score'] = amcat_analysis[['English', 'Logical', 'Quant']].mean(axis=1)
amcat_analysis['avg_amcat_percent'] = (amcat_analysis['avg_amcat_score'] / 900) * 100

# Add educational background columns
edu_columns = ['10percentage', '12percentage', 'collegeGPA']
amcat_analysis[edu_columns] = amcat[edu_columns]

# Select columns for correlation analysis
corr_columns = ['10percentage', '12percentage', 'collegeGPA', 'avg_amcat_percent']

# Calculate correlation matrix
correlation_matrix = amcat_analysis[corr_columns].corr()

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt=".2f")
plt.title("Correlation Heatmap: Educational Background vs AMCAT Performance")
plt.tight_layout()
plt.show()
```

**Salary vs Specialization **

```python
def clean_specialization(row):
    specialization_map = {
        'CSE': ['computer'],
        'ECE': ['communication'],
        'IT': ['information'],
        'EEE': ['electrical', 'electro', 'telecomm', 'power', 'embedded'],
        'Mechanical Engg': ['combus', 'mech'],
        'Chemical Engg': ['polymer', 'chem'],
        'Civil Engg': ['civil', 'ceramic'],
        'Metallurgical Engg': ['metallurgical'],
        'Instrumentation Engg': ['instrument'],
        'Production Engg': ['industrial'],
        'Biomedical Engineering': ['bio'],
        'Automobiles': ['auto'],
        'Aeronautical Engg': ['aero']
    }

    row = row.lower()
    for category, keywords in specialization_map.items():
        if any(keyword in row for keyword in keywords):
            return category
    return row

# Apply the function to the 'Specialization' column
amcat['Specialization'] = amcat['Specialization'].apply(clean_specialization)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y='Specialization',x='Salary',data=amcat)
plt.title("Distribution of salary by Specialization")
plt.show()
```

* Chemical engineering has a wider spread of salary range.

*  Production and Chemical Engineers have the highest median salary
   amongst the different spcializations.

*   Aeronautical engineering and Metallurgical Engineering had the least spread of salary ranges.
